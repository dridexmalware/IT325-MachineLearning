**1. Imputation** \
Replaces `NaN` (Not a Number) values with the `mean` of the non-missing values in the column (other values aside form the mean can also be used).
\begin{equation}
    NaN_{i,j} = \frac{\sum{x_j}}{n_j}
\end{equation}

Where:
- $NaN_{i,j}$ is the cell with the missing value in the $i$-th row and $j$-th column.
- $x_j$ are the cells with non-missing values in the $j$-th column.
- $n_j$ is the number of $x_j$ instances in the $j$-th column.

<div class="alert alert-block alert-warning">âš ï¸ Make sure that you named the sample data above as <code>sample_df1</code></div>


```python
sums_j = 0
n_j = 0

for idx, row in sample_df1.iterrows():
    X = row['X']
    Y = row['Y']
    if np.isnan(Y) == False:
        sums_j += Y
        n_j += 1

avg = sums_j / n_j
print(f"Mean: {avg}")
```

```python
for idx, row in sample_df1.iterrows():
    X = row['X']
    Y = row['Y']
    if np.isnan(Y) == True:
        sample_df1.loc[idx, 'Y'] = avg

sample_df1
```

## `Perform Data Cleaning for the Titanic Dataset`

Imputation is preferred over dropping missing values when dealing with the Titanic dataset, for a number of reasons. For one, every data point in the Titanic dataset has historic significance since each of them represents a passenger or crew member who had their own unique attributes and stories. Removing rows that have some information missing might result in loss of valuable insights about the demographics, social status and patterns of survival on board the Titanic. Secondly, the size of this dataset is not exceedingly big hence it is important to have as much data preserved as possible in order to undertake an exhaustive analysis. Imputation on the other hand can preserve data integrity while minimizing bias that could arise from simply eliminating incomplete records by means of properly chosen statistical techniques grounded on context-specific understanding. Depending on the kind of missingness involved, methods like mean, median or mode imputation for real numbers or sophisticated methods such as regression imputation and multiple imputations for both numerical and categorical variables can be employed. Furthermore, whether or not there is randomness associated with missingness patterns should guide which method would be most appropriate for imputation in order to ensure robust analysis. To sum up, I prefer using imputations which enables me keep up with a lot of data thus making it more detailed

```python
df.fillna({
    'Age': df['Age'].median(),
    'Embarked': df['Embarked'].mode()[0],
    'Cabin': 'Unknown'
}, inplace=True)

df
```

### `Should we Perform Scalar Imputation?`

*Any assumptions?*

Whenever to do scalar imputation on the Titanic dataset relies on comprehending the sort of missing data and considering these objectives. If the missing values are randomly distributed throughout the data without any bias toward a specific subset, then scalar imputation is one way to sustain integrity of the dataset without introducing large distortions. On numeric columns such as Age, median or mean can be used for substitution of missing values while mode can give a good approximation for categorical data like Embarked. Nevertheless, this technique assumes that other variables in a model donâ€™t affect the missingness, which could potentially bias results. In situations where the information loss is not at random (MNAR), scalar imputation might cause systematic error. Consequently, though I am inclined to scalar imputation due to its simplicity and efficacy in certain cases, I must also consider its limitations and carefully evaluate what kind of missing data it represents before moving forward with my analysis. By paying close attention to this factor, it will help me keep my analysis within range so that it remains as accurate and reliable as possible given that constraints have been set by datasetâ€™s conditions.

### `Perform Bootstrap Sampling Method`

<center><img src="lab1_figs/bootstrap.png" width="600"></center>

*set the random seed to 42* \
*display the histogram for the resampled data.*

```python
np.random.seed(42)
n_bootstrap_samples = 1000
bootstrap_means = []

for _ in range(n_bootstrap_samples):
    bootstrap_sample = np.random.choice(df['Age'], size=len(df['Age']), replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))

plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Bootstrap Sample Means (Age)')
plt.xlabel('Mean Age')
plt.ylabel('Frequency')
plt.show()
```

### `Treat Missing values for Cabin Variable`

### `Perform Bootstrap Sampling for Cabin`

```python
np.random.seed(42)
n_bootstrap_samples = 1000
proportions_unknown = []

for _ in range(n_bootstrap_samples):
    bootstrap_sample = df.sample(n=len(df), replace=True)
    proportion_unknown = (bootstrap_sample['Cabin'] == 'Unknown').mean()
    proportions_unknown.append(proportion_unknown)

plt.figure(figsize=(10, 6))
plt.hist(proportions_unknown, bins=30, color='skyblue', edgecolor='black')
plt.title('Bootstrap Distribution of Proportion of Unknown Cabin Allocations')
plt.xlabel('Proportion of Unknown Cabins')
plt.ylabel('Frequency')
plt.show()
```
